#Why choose Matplotlib and Seaborn?
Both libraries are among the most popular choices for data visualzation in Python Here's why:
Matlotlib: Known for its flexibility, Matplotlib allows detailed customization of visual elemnts, making it perfect For tailored plots.

Seaborn: Built on top Matplotlib, seaborn simplifies visualization tasks by offering high-level functions to create aesthetically pleasing statistical graphics.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Line plot
x =  np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x =  np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="sine wave")
plt.title(" Sine wave line plot")
plt.xlabel(" value of x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

```python
#bar chart
categories =['A', 'B', 'C']
values = [50, 100, 75]
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title("Bar chart example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class_interval = np.linspace(200, 800, 10)

data = np.random.randint(200, 800, 100)

plt.hist(data, bins=class_interval, edgecolor='black')

plt.title("Histogram Example")
plt.xlabel("Class Interval")
plt.ylabel("Frequency")

plt.show()
```

```python
#histogram and KDE plot
data = np.random.randn((1000))
sns.histplot(data, kde=True, color='green')
plt.title('Histogram with KDE Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#heatmap
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot=True, cmap="coolwarm")
plt.title("heatmap example")
plt.show()
```

```python
#heatmap
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot=True, cmap="viridis")
plt.title("heatmap example")
plt.show()
```

```python
#pair plot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show()
```
