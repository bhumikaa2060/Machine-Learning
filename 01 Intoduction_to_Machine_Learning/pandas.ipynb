###Data merge, Concatenate and Join

```python
import pandas as pd
df1 = pd.read_csv("/content/drive/MyDrive/Machine Learning/data cleaning/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/Machine Learning/data cleaning/LOTR 2.csv")
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='right', on=['FellowshipID', 'FirstName'])
```

```python
#concatenate df1 and df2
result = pd.concat([df1, df2])
result
```
